```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
np.set_printoptions(suppress=True)
scaler=StandardScaler()
```

```python
def cost(X,Y,theta):
    totalcost = 0
    M,N = X.shape
    for i in range(M):
        b =[]
        x = X[i,:]
        y = Y[i,]
        for j in range(N):
            b.append(theta[j]*x[j])
        totalcost += (1/M)*((y - sum(b))**2)
    return totalcost
```

```python
def step_grad(X,Y,alpha,theta):
    theta_slope = np.zeros(X.shape[1])
    M,N= X.shape
    for i in range(M):
        x = X[i,:]
        y = Y[i,]
        for j in range(N):
            temp = []
            for k in range(N):
                temp.append(x[k]*theta[k])
            theta_slope[j]+=(-2/M)*(y - sum(temp) )*x[j]
    new_theta = theta - (alpha*theta_slope)
    return new_theta
```

```python
def gd(X,Y, alpha, num):
    M,N = X.shape
    theta = np.zeros(N)
    for i in range(num):
        theta = step_grad(X,Y,alpha,theta)
        print(i,'Cost:',cost(X,Y,theta))
    return theta
```

```python
data = np.genfromtxt('training_ccpp_x_y_train.csv',delimiter = ',')
X = data[:,:-1]
Y_train= data[:,-1]
scaler.fit(X)
X = scaler.transform(X)
X_train =np.append(X,np.ones((len(X),1)),axis = 1)
alpha = 0.1
num = 200
theta = gd((X_train), Y_train, alpha,num)
```

```python
def predict(X_test, theta):
    M,N = X_test.shape
    Y_predict = np.zeros((M,1))
    for i in range(M):
        b = 0
        x =X_test[i]
        for j in range(N):
            b+=(x[j]*theta[j])
        Y_predict[i] = b
    return Y_predict
```

```python
data2 = np.genfromtxt('test_ccpp_x_test.csv',delimiter = ',')
X = scaler.transform(data2)
X = np.append(X,np.ones((len(X),1)),axis =1 )
Y_predict = predict(X, theta)
for i in Y_predict:
    print(i)
```
